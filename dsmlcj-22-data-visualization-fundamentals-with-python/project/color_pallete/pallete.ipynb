```python
import matplotlib.pyplot as plt

days = ['Mon', 'Tue', 'Wed']
predictions = [10.5, 12.0, 11.2]
uncertainty = [0.5, 1.2, 0.8]  # наприклад, +/- відхилення

plt.errorbar(days, predictions, yerr=uncertainty, fmt='o', capsize=5, label='Predicted Value')
plt.ylabel('Value')
plt.title('Prediction with Uncertainty')
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y_pred = np.sin(x)
error = 0.2  # 95% інтервал

plt.plot(x, y_pred, label='Prediction')
plt.fill_between(x, y_pred - error, y_pred + error, color='gray', alpha=0.2, label='95% CI')
plt.legend()
plt.show()
```
