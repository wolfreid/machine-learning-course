```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5 * np.pi, 2000)
y = np.sin(x)

fig, axs = plt.subplots(nrows=2, ncols=2,
                        subplot_kw = dict(polar = False))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)

fig.suptitle('matplotlib.pyplot.subplots() Example')
plt.show()
```
